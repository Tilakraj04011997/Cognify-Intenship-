### **Task-03**

```python
import plotly.express as px

fig = px.scatter_mapbox(df, lat="Latitude", lon="Longitude")
fig.update_layout(mapbox_style="open-street-map", mapbox_zoom=0, mapbox_center_lat = 12.9716, mapbox_center_lon = 77.5946)
fig.show()
```

```python
correlation = df[["Latitude", "Longitude", "Aggregate rating"  ]].corr()
```

```python
sns.heatmap(correlation, annot=True, cmap="coolwarm" )
```

```python
# Calculate the correlation coefficient between the restaurant's location and rating
corr = df['Latitude'].corr(df['Aggregate rating'])

# Visualize the correlation using a scatter plot
sns.scatterplot(x='Latitude', y='Aggregate rating', data=df)

# Show the plot
plt.show()

# Interpret the correlation coefficient
if corr > 0:
    print("There is a positive correlation between the restaurant's location and rating.")
elif corr < 0:
    print("There is a negative correlation between the restaurant's location and rating.")
else:
    print("There is no correlation between the restaurant's location and rating.")
```

```python
# Calculate the correlation coefficient between the restaurant's location and rating
corr = df['Longitude'].corr(df['Aggregate rating'])

# Visualize the correlation using a scatter plot
sns.scatterplot(x='Longitude', y='Aggregate rating', data=df)

# Show the plot
plt.show()

# Interpret the correlation coefficient
if corr > 0:
    print("There is a positive correlation between the restaurant's location and rating.")
elif corr < 0:
    print("There is a negative correlation between the restaurant's location and rating.")
else:
    print("There is no correlation between the restaurant's location and rating.")
```
